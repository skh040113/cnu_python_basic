#### 1) if문
##### ① 기본 구조
<p>● <span style='background-color: #fff5b1'>변숫값을 먼저 설정</span></p>
<p>● if 조건문<span style='background-color: #fff5b1'>:</span> <span style="color:red">(※ if문에서 '조건문'은 참과 거짓을 판단하는 문장)</span></p>
<p>● <span style='background-color: #fff5b1'>(들여쓰기)</span> 수행할 문장 <span style="color:red">(※ 들여쓰기는 일정하게 같은 너비로 해야 함.)</span></p>



##### ② 비교 연산자



```python
score=95

if score>=90: #점수가 90점 이상이면
    print("축하합니다. 당신은 합격입니다.") #위의 조건을 만족하면"축하합니다. 당신은 합격입니다."가 출력됨.
```

```python
score=82

if score>=90:
    print("축하합니다. 당신은 합격입니다")
```

```python
a=3
b=2

if a>b: #a가 b보다 크다면
    print("True") #위의 조건을 만족하면"True"를 출력하시오.
```

#### 2) if-else문
##### ① 기본 구조
<p>● <span style='background-color: #fff5b1'>변숫값을 먼저 설정</span></p>
<p>● if 조건문<span style='background-color: #fff5b1'>:</span></p>
<p>● <span style='background-color: #fff5b1'>(들여쓰기)</span> 수행할 문장</p>
<p>● else 조건문<span style='background-color: #fff5b1'>:</span> <span style="color:red">(※ if문 없이 독립적으로 사용할 수 없음.)</span></p>
<p>● <span style='background-color: #fff5b1'>(들여쓰기)</span> 수행할 문장</p>



```python
money=True #True는 '참(True)'를 나타내는 불(bool) 자료형으로, 파이썬 예약어임. 첫 문자를 반드시 대문자로 써야 함. 교재 102쪽 참고

if money: #a가 b보다 크다면
    print("택시를 타고 가라.") #위의 조건을 만족하면"True"를 출력하시오.
else:
    print("걸어 가라.")
```

<span style="color:#0000ff"><strong>
    2) 아래의 문장을 프로그래밍하시오.
    </strong></span>
    


```python
money=3000

if money>=3000:
    print("택시를 타고 가라.") #위의 조건을 만족하면 "택시를 타고가라."를 출력하시오.
else: #조건을 만족하지 않으면
    print("걸어가라.")#"걸어가라"를 출력하시오
```

```python
money=2700

if money<=3000:
    print("걸어가라")
else:
    print("택시를 타고 가라")
```

##### ② 조건의 연산자
<p><span style='background-color: #fff5b1'>조건을 판단</span>하기 위해 사용하는 연산자</p>



```python
money=2000
card=True

if money>=3000 or card: #돈이 3000원 이상 있거나 카드가 있으면
    print("택시를 타고 가라.") #위의 조건을 만족하면"택시를 타고 가라."를 출력하시오.
else: #조건을 만족하지 않으지 않으면
    print("걸어 가라.") #"걸어 가라."를 출력하시오.
```

##### ③ in-not-in 조건



```python
pocket=['paper', 'cellphone', 'money']

if 'money' in pocket: #주머니 안에 돈이 있으면
    print("택시를 타고 가라.") #위의 조건을 만족하면"택시를 타고 가라."를 출력하시오.
else: #조건을 만족하지 않으지 않으면
    print("걸어 가라.") #"걸어 가라."를 출력하시오.
```

```python
pocket=['paper', 'cellphone', 'money']

if 'money' in pocket: #주머니 안에 돈이 있으면
    pass #위의 조건을 만족하면 아무런 일을 하지 않도록 설정
else: #조건을 만족하지 않으지 않으면
    print("카드를 꺼내라.") #"카드를 꺼내라."를 출력하시오.
```

#### 3) if-elif-else문
##### ① 기능
<p>● 복잡함을 해결하기 위해 <span style='background-color: #fff5b1'>다중 조건 판단을 가능하게 함.</span></p>

##### ② 기본 구조
<p>● <span style='background-color: #fff5b1'>변숫값을 먼저 설정</span></p>
<p>● if 조건문<span style='background-color: #fff5b1'>:</span></p>
<p>● <span style='background-color: #fff5b1'>(들여쓰기)</span> 수행할 문장</p>
<p>● elif 조건문<span style='background-color: #fff5b1'>:</span> <span style="color:red">(※ elif는 개수에 제한 없이 사용할 수 있음.)</span></p>
<p>● <span style='background-color: #fff5b1'>(들여쓰기)</span> 수행할 문장</p>
<p>● else 조건문<span style='background-color: #fff5b1'>:</span></p>
<p>● <span style='background-color: #fff5b1'>(들여쓰기)</span> 수행할 문장</p>



##### 3) 특징
<p>● <span style='background-color: #fff5b1'>이전 조건문이 거짓일 때 수행됨.</span></p>

```python
pocket=['paper', 'cellphone']
card=False

if 'money' in pocket: #주머니 안에 돈이 있으면
    print("택시를 타고 가라.") #위의 조건을 만족하면"택시를 타고 가라."를 출력하시오.
elif card: #주머니 안에 카드가 있으면
    print("택시를 타고 가라.") #위의 조건을 만족하면"택시를 타고 가라."를 출력하시오.
else: #조건을 만족하지 않으지 않으면
    print("걸어 가라.") #"걸어 가라."를 출력하시오.
```

<span style="color:#0000ff"><strong>
    5) 아래의 조건을 만족하도록 프로그래밍하시오.
    </strong></span>
    


```python
x=10

if x == 10:
    print('10입니다.')
elif x==20:
    print('20입니다.')
else:
    print('10도 20도 아닙니다.')
```

### 3. input() 활용하기
<p> ● 사용자가 입력한 값을 변수에 대입하고 싶을 때 사용하는 기능</p>
<p> ● input은 입력되는 모든 것을 문자열로 취급함.</p>

<span style="color:#0000ff"><strong>
    연습) x의 값이 10이면 '10입니다.', x의 값이 20이면 '20입니다.'가 출력되도록 프로그래밍을 한 뒤 10을 입력해 결과를 확인하시오.
    </strong></span>

```python
#기존 방법
x=30

if x==10:
    print("10입니다.")
elif x==20:
    print("20입니다.")
else:
    print("10도 20도 아닙니다.")
```

```python
x=int(input())

if x==10:
    print("10입니다.")
elif x==20:
    print("20입니다.")
else:
    print("10도 20도 아닙니다.")
```
