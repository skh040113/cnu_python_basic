### 2. WHILE문 이해하기

<p>● 반복 범위 없이 <span style='background-color: #fff5b1'>조건에 따라 반복 수행 여부를 결정하는 경우에 사용하는 반복문</span></p>
<p>● 조건문이 ‘참’인 동안에 while문 아래의 문장이 반복해서 수행됨.</p>



#### 1) 기본 구조
<p>변수=변숫값(일반적으로 0; 변숫값 초기화)</p>
<p>while 조건식: <span style="color:red">※ 조건식을 쓰고 뒤에 반드시 쌍점(:)을 붙여야 함.</span></p>

    <p>(들여쓰기) 변화식 #변수 값의 증가/감소

#### 2) 특징
<p>● <span style='background-color: #fff5b1'>조건문으로만 동작함.</span></p>

#### 4) if문과 함께 사용하기
<p>● while문 뒤에 <span style='background-color: #fff5b1'>if문을 추가</span>하여 조건을 설정할 수 있음.</p>

```python
i=0

while i<5: #i가 5보다 작을 때 반복
    i+=1 #i가 1씩 증가
    print(i)
    
    if i==5: #i가 5와 같을 떼
        print('i가 5가 되었으므로 종료합니다.')
```

### 3. WHILE문 제어하기


#### 1) break: 반복문 빠져나가기
<p>● 제어 흐름(반복)을 중단하고 빠져나옴.</p>
<p>● while문은 조건문이 참인 동안 계속해서 내용을 반복하는데 <span style='background-color: #fff5b1'>반복을 강제로 멈추고 싶을 때 사용</span></p>
<p>● break는 <span style='background-color: #fff5b1'>일반적으로 if문과 함께 사용</span></p>



```python
i=0

while True: #무한 루프
    i+=1 #i를 1씩 증가시킴

    if i>3: #i가 3보다 클 때
        break #반복문을 끝냄. while문의 제어 흐름을 벗어남.
    print(i)
```

#### 2) continue: 반복문 처음으로 돌아가기
<p>● 제어 흐름(반복)을 유지한 상태에서 코드 실행만 건너뜀.</p>
<p>● while문의 입력 조건과 맞지 않아 <span style='background-color: #fff5b1'>처음으로 돌아가고 싶을 때 사용</span></p>
<p>● 반복문 안에서 continue를 만나면 반복문으로 처음으로 돌아가 다음 반복을 진행함.</p>



```python
i=0

while i<10: #i가 10보다 작을 때 반복
    i+=1 #i를 1씩 증가시킴
    
    if i%2==0: #i를 2로 나누었을 때 나머지가 0이면 짝수
        continue #아래 코드를 실행하지 않고 건너뜀.
    print(i)
```
