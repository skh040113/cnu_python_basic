### 1. 튜플(Tuple)

<p>● <span style='background-color: #fff5b1'>여러 데이터를 하나로 묶는</span> 자료형</p>
<p>● 몇 가지 점을 제외하면 리스트와 거의 비슷함.</p>
<p> -리스트는 []으로 요솟값을 둘러싸지만 튜플은 ()로 둘러쌈.</p>
<p> -리스트는 그 값의 생성·삭제·수정이 가능하지만 튜플은 불가능함.</p>



#### 1) 튜플 생성하기
● <span style='background-color: #fff5b1'>튜플명=(요소 1, 요소 2, 요소 3)</span>

```python
a=() #비어 있는 리스트는 a=tuple()로도 생성할 수 있음.
b=(1,) #튜플이 1개의 요소만 가질 때에는 반드시 요소 뒤에 콤마(,)를 붙여야 함.
c=(1, 2, 3)
d=1, 2, 3 #괄호()를 생략해도 튜플을 만들 수 있음.
e=('a', 'b', ('ab', 'cd')) #튜플 안에 튜플을 요솟값으로 설정할 수 있음.

print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
a=() #비어 있는 리스트는 a=tuple()로도 생성할 수 있음.
b=(1,) #튜플이 1개의 요소만 가질 때에는 반드시 요소 뒤에 콤마(,)를 붙여야 함.
c=(1, 2, 3)
d=1, 2, 3 #괄호()를 생략해도 튜플을 만들 수 있음.
e=('a', 'b', ('ab', 'cd')) #튜플 안에 튜플을 요솟값으로 설정할 수 있음.

print(type(a)) #자료형 확인하기
print(type(b)) #자료형 확인하기
print(type(c)) #자료형 확인하기
print(type(d)) #자료형 확인하기
print(type(e)) #자료형 확인하기
```

#### 2) 튜플 수정하기
● 튜플의 요솟값은 <span style='background-color: #fff5b1'>한 번 정해지면 지우거나 변경할 수 없음.</span>



```python
a=(1, 2, 'a', 'b')

del a[0] #튜플 a의 첫 번째 요소를 삭제하라는 의미

print(a) #튜플은 요솟값 삭제가 불가능함.
```

```python
a=(1, 2, 'a', 'b')

a[0]='c' #튜플 a의 첫 번째 요소를 c로 바꾸라는 의미

print(a) #튜플은 요솟값 변경이 불가능함.
```

### 2. 튜플(Tuple) 다루기
#### 1) 인덱싱(Indexing)
<p>● 문자열, 리스트와 같이 인덱싱이 가능함.</p>

```python
a=(1, 2, 'a', 'b')

print(a[1])
```

```python
a=(1, 2, 'a', 'b')

print(a[3]) #튜플 a에서 인덱스 번호 3번의 요소를 인덱싱하라는 의미
```

#### 2) 슬라이싱(Slicing)
<p>● 문자열, 리스트와 같이 슬라이싱이 가능함.</p>

```python
a=(1, 2, 'a', 'b')

print(a[1:]) #a[1]부터 튜플의 마지막 요소까지 슬라이싱하라는 의미
```

#### 3) 더하기
<p>● 문자열, 리스트와 같이 더하기가 가능함.</p>

```python
a=(1, 2)
b=(3, 4)

print(a+b) #튜플의 특정 요솟값을 바꾸는 것이 아니기 때문에 가능함.
```

#### 4) 더하기
<p>● 문자열, 리스트와 같이 곱하기가 가능함.</p>

```python
a=(1, 2)

print(a*3) #튜플의 특정 요솟값을 바꾸는 것이 아니기 때문에 가능함.
```

#### 5) 튜플 길이 구하기
<p>● 문자열, 리스트와 같이 길이 구하기가 가능함.</p>
<p>● len() 사용</p>

```python
a=(1, 2, 'a', 'b', 1)

print(len(a)) #튜플의 특정 요솟값을 바꾸는 것이 아니기 때문에 가능함.
print(a.count(1))
```
