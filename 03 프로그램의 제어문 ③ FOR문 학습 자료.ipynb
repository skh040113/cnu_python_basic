### 2. FOR문 이해하기

<p>● <span style='background-color: #fff5b1'>주어진 자료를 가져와 처리하는 경우(리스트, 튜플 등)</span> 또는 <span style='background-color: #fff5b1'>반복할 횟수가 정해져 있는 경우</span>에 사용하는 반복문</p>



#### 1) 주어진 자료로 반복하기
##### (1) 기본 구조
<p>● 리스트나 튜플, 문자열의 첫 번째 요소부터 마지막 요소까지 <span style='background-color: #fff5b1'>차례로 변수에 대입</span>하여 '수행할 문장 1…n'이 수행됨.
<p>변수=변숫값</p>
<p>for 변수 in 리스트(또는 튜플, 문자열): <span style="color:red">※ 뒤에 반드시 쌍점(:)을 붙여야 함.</span></p>
    <p>(들여쓰기) 반복 수행할 문장 입력</p>
    


```python
a=['one', 'two', 'three']

for i in a: #리스트 a에 들어 있는 요소들을 변수 i에 순서대로 저장
    print(i)
```

```python
a='가나다라'

for i in a: #a에 들어 있는 요소들을 변수 i에 순서대로 저장
    print(i)
```

```python
a=[(1, 2), (3, 4), (5, 6)]

for (first, last) in a: #리스트 a에 들어 있는 요소들을 변수 i에 순서대로 저장
    print(first, last)
```

<span style="color:#0000ff"><strong>
    1) for문을 사용하여 튜플 a의 요소를 차례대로 출력하시오.</strong>

```python
a=('apple', 'orange', 'banana')

for i in a:
    print(i)
```

<span style="color:#0000ff"><strong>
    2) for문을 사용하여 리스트 a에 들어 있는 튜플들을 각각 더한 값을 아래와 같이 출력하시오.</strong>
    


```python
a=[(1, 2), (3, 4), (5, 6)]

for (first, last) in a:
    print(first+last)
```

<span style="color:#0000ff"><strong>
    3) for문을 사용하여 Python의 문자를 하나씩 분리하여 출력하시오.</strong>

```python
for i in "Python":
    print(i, end=' ') # end=' '을 지정하면 줄바꿈을 하지 않고 출력
```

##### (2) 응용하기
<p>① <span style='background-color: #fff5b1'>reversed()</span>를 사용하여 자료형을 순서를 뒤집기.</p>

```python
a='Python'

for i in reversed(a):
    print(i)
```

```python
a='Python'

for i in reversed(a):
    print(i, end=' ') #end=' '을 지정하면 줄바꿈을 하지 않고 각 문자 사이에 공백을 두어 출력할 수 있게 도와 줌.
```

<span style="color:#0000ff"><strong>
    4) for문을 사용하여 Python의 문자를 뒤집어서 하나씩 분리하여 출력하시오. (단, 별도의 변수를 지정하지 않음.)</strong>

② <span style='background-color: #fff5b1'>if-else문을 사용</span>하여 조건을 제시하기.</p>



```python
a=[91, 26, 68, 46, 81]

for i in a:
    
    if i>=60:
        print(i, '합격입니다.')
        
    else:
        print(i, '불합격입니다.')
```

<span style="color:#0000ff"><strong>
    5) for문을 사용하여 아래와 같은 결과가 출력될 수 있도록 하시오.</strong>
    


```python
a=[91, 26, 68, 46, 81]

for i in a:
    
    if i>=60:
        print(str(i)+"점, 합격입니다.")
        
    else:
        print(str(i)+"점, 합격입니다.")
```

```python
a=[91, 26, 68, 46, 81]

for i in a:
    
    if i>=60:
        print("%d점, 합격입니다." %i)
        
    else:
        print("%d점, 불합격입니다." %i)
```

<p>③ <span style='background-color: #fff5b1'>continue를 사용</span>하기.</p>
<p>● for문 안의 문장을 수행하는 도중에 continue를 만나면 for문의 처음으로 돌아감.</p>



```python
a=[91, 26, 68, 46, 81]
number=0

for i in a:
    number+=1
    
    if i<60:
        continue
        
    print('%d번 학생 축하합니다. 합격입니다.' % number)
```

#### 2) 주어진 횟수로 반복하기
##### (1) 기본 구조
<p>● 숫자 리스트를 자동으로 만들어 주는 range()와 함께 사용함.</p>
<p>변수=변숫값</p>
<p>for 변수 in range(횟수): <span style="color:red">※ 뒤에 반드시 쌍점(:)을 붙여야 함.</span></p>
    <p>(들여쓰기) 반복 수행할 문장 입력</p>
<p>● range()에 숫자(N)를 입력하면 0부터 N-1까지 정수를 순서대로 생성함.</p>



```python
for i in range(5): #변수 i에 아래의 문장을 5번 반복함.
    print('Hello Python!')
```

```python
for i in range(5): #변수 i에 순서대로 0, 1, 2, 3, 4를 넣음.
    print(i)
```

##### (2) 응용하기
<p>① range()에서 <span style='background-color: #fff5b1'>출력 범위를 지정</span>할 수 있음.</p>
<p>변수=변숫값</p>
<p>for 변수 in range(<span style='background-color: #fff5b1'>시작 숫자, 끝 숫자</span>): <span style="color:red">※ 뒤에 반드시 쌍점(:)을 붙여야 함.</span> <span style="color:#0000ff">(※ 범위를 지정할 때에는 슬라이싱의 원리를 생각할 것!)</span></p>
    <p>(들여쓰기) 반복 수행할 문장 입력</p>

```python
for i in range(1, 5): #변수 i에 순서대로 1부터 5 미만의 숫자를 넣음.
    print(i)
```

<span style="color:#0000ff"><strong>
    7) range()를 사용하여 아래와 같은 결과가 출력되게 하시오.</strong>



<p>② range()에서 <span style='background-color: #fff5b1'>증가/감소의 폭을 설정</span>하여 출력 범위를 지정할 수 있음.</p>
<p>변수=변숫값</p>
<p>for 변수 in range(시작 숫자, 끝 숫자, <span style='background-color: #fff5b1'>증감의 폭</span>): <span style="color:red">※ 뒤에 반드시 쌍점(:)을 붙여야 함.</span> <span style="color:#0000ff">(※ 증감의 폭을 지정할 때에는 연산 기호를 생각할 것!)</span></p>
    <p>(들여쓰기) 반복 수행할 문장 입력</p>

```python
for i in range(1, 11, 2): #변수 i에 순서대로 1부터 10을 넣되 증가 폭을 2로 함.
    print(i)
```

<span style="color:#0000ff"><strong>

<p>③ <span style='background-color: #fff5b1'>reversed()를 사용해</span> 숫자를 역순으로 출력할 수 있음.</p>
<p>● reversed(range(횟수))</p>
<p>변수=변숫값</p>
<p>for 변수 in reversed(횟수): <span style="color:red">※ 뒤에 반드시 쌍점(:)을 붙여야 함.</span></p>
    <p>(들여쓰기) 반복 수행할 문장 입력</p>

```python
for i in reversed(range(5)): #0부터 5 미만까지의 숫자를 역순으로
    print(i)
```

<p>● reversed(시작 숫자, 끝 숫자)</p>
<p>변수=변숫값</p>
<p>for 변수 in reversed(시작 숫자, 끝 숫자)): <span style="color:red">※ 뒤에 반드시 쌍점(:)을 붙여야 함.</span></p>
    <p>(들여쓰기) 반복 수행할 문장 입력</p>

```python
for i in reversed(range(0, 10)): #1부터 10 미만까지의 숫자를 역순으로
    print(i)
```

<p>● reversed(range(시작 숫자, 끝 숫자, 증가폭))</p>
<p>변수=변숫값</p>
<p>for 변수 in reversed(range(시작 숫자, 끝 숫자, 증가폭)): <span style="color:red">※ 뒤에 반드시 쌍점(:)을 붙여야 함.</span></p>
    <p>(들여쓰기) 반복 수행할 문장 입력</p>

```python
for i in reversed(range(0, 10, 2)): #1부터 10 미만까지의 숫자를 역순으로
    print(i)
```

### 3. 리스트 내포 사용하기
<p>● 리스트 안에 for문을 포함하는 것을 리스트 내포(List comprehension)라고 함.</p>

```python
a=[1, 2, 3, 4]
result=[]
for num in a:
    result.append(num*3)
    
print(result)
```

```python
a=[1, 2, 3, 4]
result=[num*3 for num in a]

print(result)
```
