#### 1) 정의

#### 2) 리스트의 생성 방법
● <span style='background-color: #fff5b1'>리스트명=[요소 1, 요소 2, 요소 3]</span>

```python
a=[] #비어 있는 리스트는 a=list()로도 생성할 수 있음.
b=[1, 2, 3] #각 요소의 값은 쉼표(,)로 구분
c=['Life', 'is', 'too', 'short'] #문자열을 요소로 지정할 때에는 따옴표를 사용
d=[1, 2, 'Life', 'is'] #숫자형, 문자열을 섞어서 리스트의 요소로 지정할 수 있음.
e=[1, 2, ['Life', 'is']] #리스트를 요솟갑승로 가질 수 있음.

print(a)
print(b)
print(c)
print(d)
print(e)
```

#### 3) 리스트의 인덱싱(Indexing)
<p>● 리스트도 문자열처럼 인덱싱이 가능함.</p>
<p>● <span style='background-color: #fff5b1'>인덱스 번호는 0부터 시작</span>한다는 것을 꼭 기억하기!</p>
<p>● <span style='background-color: #fff5b1'>리스트명[인덱스 번호]</span></p>

#### ① 인덱싱
<p>● 리스트의 요솟값이 숫자인 경우</p>

```python
a=[1, 2, 3] #변숫값으로 리스트를 설정

print(a[1]) #변수 a의 값 중 두 번째 요소를 구하려면 인덱스 번호 1 또는 -2를 입력
print(a[-2])
```

<p>● 리스트의 요솟값이 숫자와 문자 혼합형인 경우</p>

```python
a=[1, 2, 3, ['a', 'b', 'c']]

print(a[3])
```

<p>● 응용: 다중 리스트에서 인덱싱하기</p>

```python
a=[1, 2, 3, ['a', 'b', 'c']]

print(a[3][0]) #네 번째 요솟값에서 a만을 출력하고 싶다면?

print(a[3][2])
```

<span style="color:#0000ff"><strong>

    </strong></span>

```python
a=[1, 2, ['a', 'b', ['Life', 'is']]]



print(a[2][2][0])
```

##### ② 사칙연산

```python
a=[1, 2, 3]
b=a[0]
c=a[1]

print(a[0]+a[1]) #변수 a의 리스트 요솟값 중 첫 번째(1)와 두 번째 요소(2)의 값을 더할 때 인덱싱을 활용
print(b+c)
```

#### 4) 리스트의 슬라이싱(Slicing)
<p>● 리스트도 문자열처럼 슬라이싱이 가능함.</p>
<p>● <span style='background-color: #fff5b1'>인덱스 번호는 0부터 시작</span>한다는 것을 꼭 기억하기!</p>
<p>● <span style='background-color: #fff5b1'>리스트명[시작 번호:끝 번호]</span></p>

```python
a=[1, 2, 3, 4, 5] #리스트의 요솟값 중, 1과 2만을 출력하고 싶다면?

print(a[0:2]) #범위를 지정할 때 끝 번호는 구하고 싶은 위치+1을 해야 함.
```

```python
a=[1, 2, 3, 4, 5]
b=a[:2] #리스트의 요솟값 중, 1과 2만을 b의 변숫값으로 설정
c=a[2:] #리스트의 요솟값 중, 3, 4, 5만을 b의 변숫값으로 설정

print(b)
print(c)
```

<span style="color:#0000ff"><strong>
    [연습] 아래의 리스트에서 슬라이싱 기법을 사용해 리스트 [2, 3]을 만드시오.
    </strong></span>

```python
a=[1, 2, 3, 4, 5]

print(a[1:3])
```

#### 6) 리스트의 수정과 삭제
#### ① 수정하기

```python
a=[1, 2, 3]

a[2]=4 #리스트 a의 세 번째 값을 4로 변환


print(a) #수정한 리스트 값을 출력할 수 있음.
```

```python
a=[1, 2, 3]

a[0]=5

print(a)
```

#### ② 삭제하기
<p>● de 함수를 사용하여 <span style='background-color: #fff5b1'>특정 요소를 삭제할 수 있음.</span></p>
<p>● <span style='background-color: #fff5b1'>del 객체</span>(이때 객체는 '리스트명[인덱스 번호]'와 같이 작성)</p>

```python
a=[1, 2, 3]
del a[1] #리스트 a에서 첫 번째 요솟값을 삭제

print(a) #첫 번째 요솟값을 삭제한 상태로 리스트 a 출력
```

● 리스트의 여러 요소를 한꺼번에 삭제하기

```python
a=[1, 2, 3, 4, 5]
del a[2:] #리스트 a에서 세 번째 요솟값부터 마지막 요솟값까지 삭제

print(a) #여러 요솟값을 삭제한 상태로 리스트 a 출력
```

<span style="color:#0000ff"><strong>
    [연습] 아래의 리스트 a에서 두 번째부터 네 번째 요솟값을 삭제한 후 리스트 a의 다섯 번째 요솟값을 6으로 수정하시오.
    </strong></span>

```python
a=[1, 2, 3, 4, 5]

del a[1:4]

a[1]=6

print(a)
```

```python
a=[1, 2, 3, 4, 5]

a[4]=6
 
del a[1:4]

print(a)
```
