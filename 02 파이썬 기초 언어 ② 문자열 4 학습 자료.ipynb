<p><strong>② 위치 알려주기 1</strong>: find()</p>

```python
a='Python is the best choice' #변수 a에 Python is the best choice를 변수값으로 지정

print(a.find('b')) #변수 a의 값 중에 문자 b가 맨 처음 나온 위치(인덱스 번호)를 알고 싶다면?
```

<p><strong>⑦ 왼쪽 공백 지우기</strong>: lstrip()</p>

```python
a='  hi.  '

print(a.lstrip()) #문자열 왼쪽에 있는 한 칸 이상의 연속된 모든 공백들을 삭제
```

<p><strong>⑩ 문자열 바꾸기</strong>: replace()</p>

```python
a='Life is too short.'

print(a.replace('Life', 'Your leg')) #replace(바뀔 문자열, 바꿀 문자열)
```

<p><strong>⑪ 문자열 나누기</strong>: split()</p>

```python
a='Life is too short.'

print(a.split()) #공백을 기준으로 문자열을 나눔.
```

```python
a='a:b:c:d'

print(a.split(':')) #괄호 안에 특정 값이 있으면 괄호 안의 값을 기준으로 나눔.
```
