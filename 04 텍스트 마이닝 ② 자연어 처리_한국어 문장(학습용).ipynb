### 1. 자연어 처리(NLP; Natural Language Processing)란?

#### 1) 정의
<p>● Software를 사용하여 <span style='background-color: #fff5b1'>인간의 언어</span>를 처리·분석하는 기술.</p>

#### 2) 활용 분야
<p>● 검색, 기계 번역, 챗봇, 음성 인식 등.</p>
<p>● 맞춤법 검사, 철자 교정 등.</p>
<p>● 데이터 시각화을 수행하기 위한 데이터 분석 작업 등.</p>

#### 3) 도구

<p>● Java(JDK): 웹 애플리케이션 코딩에 사용되는 언어로, 다양한 분야에서 사용.</p>
<p>● JPype: Java와 Python을 연결하는 라이브러리.</p>
<p>● NLTK: 펜실베니아 대학 컴퓨터 정보학과의 Steven Bird, Edward Loper가 <span style='background-color: #fff5b1'>영문 자연어 처리</span>를 가능하게 하려고 개발한 모듈.</p>
<p>● KoNLPy: Lucy Park이 파이썬으로 <span style='background-color: #fff5b1'>한글 자연어 처리</span>를 가능하게 하려고 개발한 모듈.</p>

### 2. 문장 분석
#### 1) 꼬꼬마(Kkma) 모듈
<p>● 서울대학교 IDS(Intelligent Data Systems) 연구실 개발</p>
<p>● http://kkma.snu.ac.kr/</p>
<p>● 문장을 구분할 때 사용하는 <span style='background-color: #fff5b1'>sentence()</span>를 제공</p>

##### (1) 환경 구축
<span style="color:#0000ff">1단계. KoNLPy 실행</span>

```python
from konlpy.tag import Kkma
```

```python
from konlpy.tag import Kkma
```

<span style="color:#0000ff">2단계. Kkma() 설정</span>

```python
kkma=Kkma()
```

```python
kkma=Kkma()
```

##### (2) 문장 분석: sentence()
<span style="color:#0000ff">1단계. 마침표(.)가 있는 문장</span>

```python
print(kkma.sentences('안녕하세요. 제 이름은 홍길동입니다.'))
```

```python
a= '안녕하세요. 제 이름은 홍길동입니다.'

print(kkma.sentences(a))
```

<span style="color:#0000ff">2단계. 마침표(.)가 없는 문장</span>

```python
print(kkma.sentences('안녕하세요 제 이름은 홍길동입니다')) #입력한 문장 안에 마침표(.)가 없어도 별개의 문장으로 인식하고 구분함.
```

```python
print(kkma.sentences('안녕하세요 제 이름은 홍길동입니다'))
```

<span style="color:#0000ff">3단계. 명사(Noun) 추출: nouns()</span>

```python
print(kkma.nouns('안녕하세요. 제 이름은 홍길동입니다.'))
```

```python
print(kkma.nouns('안녕하세요. 제 이름은 홍길동입니다.'))
```

```python
a= '안녕하세요. 제 이름은 홍길동입니다.'
print(kkma.nouns(a))
```

<span style="color:#0000ff">4단계. 품사 확인: pos()</span>
<p>● NNG: 보통명사, NNP: 고유명사, NP: 대명사</p>
<p>● JKS: 주격조사, JX: 보조사, EFN: 평서형 종결 어미</p>
<p>● VV: 동사, VA: 형용사</p>
<p>● SP: 문장부호(쉼표, 가운뎃점, 콜론, 빗금) 등</p>

```python
print(kkma.pos('안녕하세요. 제 이름은 홍길동입니다.'))
```

#### 2) Open Korean Text 모듈
<p>● 과거 Twitter 한국어 형태소 분석기</p>
<p>● https://github.com/open-korean-text/open-korean-text</p>
<p>● 구(phrase)를 추출하는 <span style='background-color: #fff5b1'>phrases( )</span>, 문장을 정교화해 주는 <span style='background-color: #fff5b1'>normalize( )</span>를 제공</p>

##### (1) 환경 구축
<span style="color:#0000ff">1단계. KoNLPy 실행</span>

```python
from konlpy.tag import Okt
```

<span style="color:#0000ff">2단계. Okt() 설정</span>

```python
okt=Okt()
```

##### (2) 문장 분석
<span style="color:#0000ff">1단계. 명사(Noun) 추출: nouns()</span>

```python
print(okt.nouns('안녕하세요. 제 이름은 홍길동입니다.'))
```

<span style="color:#0000ff">2단계. 품사 확인: pos()</span>
<p>● Noun: 명사, Adjective: 형용사, Verb: 동사</p>
<p>● Josa: 조사, Punctuation: 문장 부호 등</p>

```python
print(okt.pos('안녕하세요. 제 이름은 홍길동입니다.'))
```

<span style="color:#0000ff">3단계. 형태소 추출: morphs()</span>

```python
print(okt.morphs('안녕하세요. 제 이름은 홍길동입니다.'))
```

<span style="color:#0000ff">4단계. 구(Phrase) 추출: phrases()</span>

```python
print(okt.phrases('안녕하세요. 제 이름은 홍길동입니다.'))
```

<span style="color:#0000ff">5단계. 문장 정규화(normalize): normalize()</span>

```python
print(okt.normalize('앜ㅋㅋㅋ 홍길동입니닼ㅋㅋ'))
```

#### 3) 한나눔(Hannanum) 모듈

<p>● KAIST Semantic Web Research Center에서 개발</p>
<p>● http://semanticweb.kaist.ac.kr/hannanum/</p>
<p>● 오픈 소스</p>
<p>● 문장은 분석하지 못함.</p>

##### (1) 환경 구축
<span style="color:#0000ff">1단계. KoNLPy 실행</span>

```python
from konlpy.tag import Hannanum
```

<span style="color:#0000ff">2단계. Hannanum() 설정</span>

```python
hannanum=Hannanum()
```

##### (2) 단어 분석
<span style="color:#0000ff">1단계. 명사(Noun) 추출: nouns()</span>

```python
print(hannanum.nouns('안녕하세요. 제 이름은 홍길동입니다.'))
```

<span style="color:#0000ff">2단계. 품사 확인: pos()</span>
<p>● N: 명사, 대명사, 수사, P: 동사, 형용사</p>
<p>● J: 격조사, 보조사, 접속조사, E: 어미</p>
<p>● S: 문장 부호</p>

```python
print(hannanum.pos('안녕하세요. 제 이름은 홍길동입니다.'))
```

<span style="color:#0000ff">3단계. 형태소 추출: morphs()</span>

```python
print(hannanum.morphs('안녕하세요. 제 이름은 홍길동입니다.'))
```

### 3. 모듈별 분석 결과 비교
#### 1) 명사(Noun) 추출

```python
print('● Kkma(kkma())')
print(kkma.nouns('안녕하세요. 제 이름은 홍길동입니다.'))
print('● 개수:', len(kkma.nouns('안녕하세요. 제 이름은 홍길동입니다.')))
print('='*50)

print('● Open Korean Text(okt())')
print(okt.nouns('안녕하세요. 제 이름은 홍길동입니다.'))
print('● 개수:', len(okt.nouns('안녕하세요. 제 이름은 홍길동입니다.')))
print('='*50)

print('● Hannanum(hannanum())')
print(hannanum.nouns('안녕하세요. 제 이름은 홍길동입니다.'))
print('● 개수:', len(hannanum.nouns('안녕하세요. 제 이름은 홍길동입니다.')))
```

#### 명사 추출은 꼬꼬마가 제일 나음 
- okt는 저+의 저의를 제라고 하나로 봄
- 한나눔은 입니까지 하나의 명사로 봄
    

#### 2) 품사 확인

```python
print('● Kkma(kkma())')
print(kkma.pos('안녕하세요. 제 이름은 홍길동입니다.'))
print('● 개수:', len(kkma.pos('안녕하세요. 제 이름은 홍길동입니다.')))
print('='*138)

print('● Open Korean Text(okt())')
print(okt.pos('안녕하세요. 제 이름은 홍길동입니다.'))
print('● 개수:', len(okt.pos('안녕하세요. 제 이름은 홍길동입니다.')))

print('='*138)

print('● Hannanum(hannanum())')
print(hannanum.pos('안녕하세요. 제 이름은 홍길동입니다.'))
print('● 개수:', len(hannanum.pos('안녕하세요. 제 이름은 홍길동입니다.')))
```

#### 3) 형태소 분석

```python
print('● Kkma(kkma())')
print(kkma.morphs('안녕하세요. 제 이름은 홍길동입니다.'))
print('● 개수:', len(kkma.morphs('안녕하세요. 제 이름은 홍길동입니다.')))
print('='*138)

print('● Open Korean Text(okt())')
print(okt.morphs('안녕하세요. 제 이름은 홍길동입니다.'))
print('● 개수:', len(okt.morphs('안녕하세요. 제 이름은 홍길동입니다.')))
print('='*138)

print('● Hannanum(hannanum())')
print(hannanum.morphs('안녕하세요. 제 이름은 홍길동입니다.'))
print('● 개수:', len(hannanum.morphs('안녕하세요. 제 이름은 홍길동입니다.')))
```
